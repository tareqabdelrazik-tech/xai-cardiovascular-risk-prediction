```python
import pandas as pd
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

```

At both Sets X_Train and X_Test

```python

# Look for the true path toward in your computer toward data and name the Dataset - HeartDisease.csv for it to work
HearDisease_csv = pd.read_csv('/home/tareqrazik/Downloads/HeartDisease.csv')

X_train, X_test= train_test_split(HearDisease_csv, test_size=0.99,random_state=48)

#The purpose is to keep all X_train , X_test split in order to make DataCleaning as clean as possible
#Note : since Our Dataset is too Big , making test_size = 0.99 will help our computers in model tuning
#Note : this needs to change at the end , since its not optimal yet it won't have an effect on our visulization
y_train ,y_test = X_train.pop('Heart_Disease') , X_test.pop('Heart_Disease')

y_train , y_test = y_train.map({'Yes':True ,'No':False}) , y_test.map({'Yes':True,'No':False})
# Drop Our Target_Variables in addition to applying trnasformation
```

Our Work only Based on Training Data - we Only learn from the training data, and keep the test data completely separate until the very end Visualization methods Only applicable on X_train , y_train Otherwise We as Our model will be Overfitting

```python
plt.figure(figsize=(10, 6))
sns.histplot(X_train, kde=True, stat='density')
plt.title('Normal Distribution - Height and Weight')
plt.xlabel('Values')
plt.ylabel('Density')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x=np.log(X_train['Height_(cm)']), y=np.log(X_train['Weight_(kg)']), 
            scatter_kws={'alpha':0.6, 'color':'blue', 's':50},
            line_kws={'color':'red', 'linewidth':2, 'label':'Regression Line'})
plt.xlabel('Log(Height) cm', fontsize=12)
plt.ylabel('Log(Weight) kg', fontsize=12)
plt.title('Height vs Weight with Regression Line', fontsize=14, fontweight='light')
plt.legend()
plt.tight_layout()
plt.show()
```

Our analysis reveals an intuitive yet important pattern: weight increases linearly with height. This relationship validates basic expectations and also confirms the quality  of our collected data.
